```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from splinter import Browser
```

```python
# Drivers`s Path
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Go to NCAA Boxscore through splinter module
score_url = 'https://www.ncaa.com/game/lacrosse-women/d1/2019/05/02/army-navy/boxscore'
browser.visit(score_url)
```

```python
html = browser.html
#response = requests.get(url)
soup = BeautifulSoup(html, 'lxml')
tables = pd.read_html(html)
away_table = tables[1]
away_table
```

```python
# Go to NCAA Play by play  through splinter module
score_url = 'https://www.ncaa.com/game/lacrosse-women/d1/2019/05/02/army-navy/play-by-play'
browser.visit(score_url)
```

```python
html = browser.html
#response = requests.get(url)
soup = BeautifulSoup(html, 'lxml')
all_rows = soup.find_all('tr')
for row in all_rows:
    try:
        print(row.find(class_="time").text)
        img = row.find(class_="team")
        print(img.find("img")["title"])
        print(row.find(class_="description").text)
        print(row.find(class_="score").text)
        
    except:
        pass

#tables = pd.read_html(html)
#away_table = tables[1]
#away_table
```
