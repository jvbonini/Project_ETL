```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
from splinter import Browser
```

```python
# Drivers`s Path
executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Go to NCAA Boxscore through splinter module
score_url = 'https://www.ncaa.com/game/lacrosse-women/d1/2019/05/02/army-navy/boxscore'
browser.visit(score_url)
```

```python
html = browser.html
#response = requests.get(url)
soup = BeautifulSoup(html, 'lxml')
tables = pd.read_html(html)
box_table = tables[1]
box_table
```

```python
  # Read the BoxScore Data and store Pandas data frame
box_table_df = box_table
```

```python
# Go to NCAA Play by play  through splinter module
score_url = 'https://www.ncaa.com/game/lacrosse-women/d1/2019/05/02/army-navy/play-by-play'
browser.visit(score_url)
html = browser.html
#response = requests.get(url)
soup = BeautifulSoup(html, 'lxml')
all_rows = soup.find_all('tr')
df_data = []
for row in all_rows:
    try:
        img = row.find(class_="team")
        df_data.append({
            "time": row.find(class_="time").text,
            "team": img.find("img")["title"],
            "score": row.find(class_="score").text,
            "description": row.find(class_="description").text
        })
        #print(row.find(class_="time").text)
        #print(img.find("img")["title"])
        #print(row.find(class_="score").text)
        #print(row.find(class_="description").text)
    except:
        pass
```

```python
new_dataframe = pd.DataFrame(df_data)
new_dataframe
```
